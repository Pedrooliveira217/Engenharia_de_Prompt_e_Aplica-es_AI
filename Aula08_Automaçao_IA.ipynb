```python
# Solicita ao usuário que digite a temperatura do sistema em Celsius
temperature_str = input("Digite a temperatura do sistema em Celsius: ")

try:
    # Tenta converter a entrada do usuário para um número decimal (float)
    temperature = float(temperature_str)

    # Verifica se a temperatura é maior que 30°C
    if temperature > 30:
        # Se for maior que 30°C, imprime um alerta de superaquecimento incluindo a temperatura atual
        print(f"ALERTA: Superaquecimento! A temperatura atual é {temperature}°C e está acima de 30°C.")
    else:
        # Se não for maior que 30°C (ou seja, 30°C ou menos), imprime que a temperatura está normal, incluindo a temperatura atual
        print(f"Temperatura normal: O sistema está funcionando dentro dos limites ideais com {temperature}°C.")
except ValueError:
    # Se ocorrer um erro ao converter a entrada para float (ex: usuário digitou texto),
    # imprime uma mensagem de erro solicitando um valor numérico
    print("Erro: Por favor, digite um valor numérico para a temperatura.")
```
