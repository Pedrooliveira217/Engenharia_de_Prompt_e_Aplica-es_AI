```python
# Solicita ao usuário que digite a temperatura do sistema em Celsius
temperature_str = input("Digite a temperatura do sistema em Celsius: ")

try:
    # Tenta converter a entrada do usuário para um número decimal (float)
    temperature = float(temperature_str)

    # Verifica se a temperatura é maior que 30°C
    if temperature > 30:
        # Se for maior que 30°C, imprime um alerta de superaquecimento incluindo a temperatura atual
        print(f"ALERTA: Superaquecimento! A temperatura atual é {temperature}°C e está acima de 30°C.")
    else:
        # Se não for maior que 30°C (ou seja, 30°C ou menos), imprime que a temperatura está normal, incluindo a temperatura atual
        print(f"Temperatura normal: O sistema está funcionando dentro dos limites ideais com {temperature}°C.")
except ValueError:
    # Se ocorrer um erro ao converter a entrada para float (ex: usuário digitou texto),
    # imprime uma mensagem de erro solicitando um valor numérico
    print("Erro: Por favor, digite um valor numérico para a temperatura.")
```

Criar um programa que organize arquivos bagunçados em pastas automaticamente, baseando-se apenas na extensão do arquivo (ex: .pdf, .jpg).

```python
import os
import shutil

def organize_files(directory_path):
    if not os.path.isdir(directory_path):
        print(f"Erro: O diretório '{directory_path}' não existe.")
        return

    print(f"Organizando arquivos em: {directory_path}")

    # Mapeamento de extensões para nomes de pastas em português
    extension_mapping = {
        "txt": "Textos",
        "pdf": "Documentos PDF",
        "doc": "Documentos Word",
        "docx": "Documentos Word",
        "xls": "Planilhas Excel",
        "xlsx": "Planilhas Excel",
        "ppt": "Apresentações PowerPoint",
        "pptx": "Apresentações PowerPoint",
        "jpg": "Imagens JPG",
        "jpeg": "Imagens JPEG",
        "png": "Imagens PNG",
        "gif": "Imagens GIF",
        "mp3": "Músicas",
        "wav": "Áudios",
        "mp4": "Vídeos",
        "avi": "Vídeos",
        "zip": "Arquivos Compactados",
        "rar": "Arquivos Compactados",
        "py": "Scripts Python",
        "js": "Scripts JavaScript",
        "html": "Páginas Web",
        "css": "Folhas de Estilo CSS",
        "json": "Arquivos JSON",
        "csv": "Arquivos CSV",
        "md": "Arquivos Markdown"
    }

    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)

        # Ignora diretórios
        if os.path.isdir(file_path):
            continue

        # Pega a extensão do arquivo
        _, file_extension = os.path.splitext(filename)

        # Remove o ponto da extensão e a converte para minúsculas
        if file_extension:
            raw_extension = file_extension[1:].lower()
            # Usa o mapeamento se a extensão estiver presente, caso contrário, usa a extensão original capitalizada
            extension_folder_name = extension_mapping.get(raw_extension, raw_extension.capitalize())
        else:
            extension_folder_name = "Sem Extensão"

        destination_folder = os.path.join(directory_path, extension_folder_name)

        # Cria a pasta de destino se ela não existir
        if not os.path.exists(destination_folder):
            os.makedirs(destination_folder)
            print(f"Pasta criada: {destination_folder}")

        # Move o arquivo para a pasta de destino
        try:
            shutil.move(file_path, destination_folder)
            print(f"Movido: '{filename}' para '{extension_folder_name}/'")
        except Exception as e:
            print(f"Erro ao mover '{filename}': {e}")

    print("Organização de arquivos concluída.")

# Exemplo de uso:
# Descomente as duas linhas abaixo e substitua o caminho para usar a função.
directory_to_organize = "/content/sample_data" # Substitua pelo caminho da pasta que você deseja organizar
organize_files(directory_to_organize)
```

Desenvolver um script que consulte uma API pública (ViaCEP) e retorne as informações de endereço formatadas de maneira legível.

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço de um determinado CEP.
    Inclui tratamento de exceções para erros de rede e CEPs inválidos.
    """
    # Remove quaisquer caracteres não numéricos do CEP
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP tem 8 dígitos
    if len(cep) != 8:
        print("Erro: O CEP deve conter 8 dígitos numéricos.")
        return

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Lança uma exceção para códigos de status HTTP de erro (4xx ou 5xx)

        data = response.json()

        if 'erro' in data and data['erro']:
            print(f"Erro: CEP '{cep}' não encontrado ou inválido.")
        else:
            print("\n--- Informações do Endereço ---")
            print(f"CEP: {data.get('cep', 'N/A')}")
            print(f"Logradouro: {data.get('logradouro', 'N/A')}")
            print(f"Complemento: {data.get('complemento', 'N/A')}")
            print(f"Bairro: {data.get('bairro', 'N/A')}")
            print(f"Localidade: {data.get('localidade', 'N/A')}")
            print(f"UF: {data.get('uf', 'N/A')}")
            print(f"DDD: {data.get('ddd', 'N/A')}")
            print("-------------------------------")

    except requests.exceptions.HTTPError as http_err:
        print(f"Erro HTTP: {http_err} - Verifique o CEP ou a disponibilidade do serviço.")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Erro de Conexão: {conn_err} - Verifique sua conexão com a internet.")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Erro de Timeout: {timeout_err} - A requisição demorou muito para responder.")
    except requests.exceptions.RequestException as req_err:
        print(f"Erro na Requisição: {req_err} - Ocorreu um erro inesperado ao fazer a requisição.")
    except ValueError:
        print("Erro: Resposta inválida da API (não é um JSON válido).")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

# Exemplo de uso:
if __name__ == "__main__":
    while True:
        cep_digitado = input("Digite o CEP (apenas números, ou 'sair' para encerrar): ")
        if cep_digitado.lower() == 'sair':
            break
        consultar_cep(cep_digitado)
```
